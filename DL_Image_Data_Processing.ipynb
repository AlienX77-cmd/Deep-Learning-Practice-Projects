##**Resizing the image using Pillow Library**

```python
from PIL import Image
```

```python
img = Image.open('/content/strategists.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('strategists_resized.jpg')
```

```python
# Displaying the image from numpy array

img_res = mpimg.imread('/content/strategists_resized.jpg')
im_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

###**Converting RGB image to Grayscale image with OpenCV**

```python
# importing the OpenCV library
import cv2
```

```python
img = cv2.imread('/content/strategists.jpg')
```

```python
type(img)
```

```python
img.shape #3 means 3 channel = RGB colors
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape #grayscale only has 1 channel
```

cv2.imshow() will display the image, but this will not be allowed in Google Colab.

use **from google.colab.patches import cv2_imshow**

```python
from google.colab.patches import cv2_imshow
```

```python
# Displaying the image
cv2_imshow(grayscale_image)
```

```python
# Saving the grayscale image
cv2.imwrite('strategists_grayscale_image.jpg', grayscale_image)
```
